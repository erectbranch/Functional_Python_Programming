## ëª…ë ¹í˜• íŒ¨ëŸ¬ë‹¤ì„ êµ¬ë¶„í•˜ê¸°

ì•„ë˜ëŠ” íŒŒì´ì¬ì˜ ê°ì²´ì§€í–¥ì  íŠ¹ì„±ì„ ë¬´ì‹œí•˜ê³  ë§Œë“  ê°„ë‹¨í•œ ì ˆì°¨í˜• ì•Œê³ ë¦¬ì¦˜ì´ë‹¤. 10ë¯¸ë§Œì˜ ìì—°ìˆ˜ì—ì„œ 3ì´ë‚˜ 5ì˜ ë°°ìˆ˜ë¥¼ ëª¨ë‘ ë”í•œ ê°’ì„ ê²°ê³¼ë¡œ ë„ì¶œí•œë‹¤.


```python
s = 0

# nì— 1ì—ì„œ 9ê¹Œì§€ í• ë‹¹
for n in range(1, 10): 
    
    # nì´ 3ìœ¼ë¡œ ë‚˜ëˆ„ì–´ì§€ê±°ë‚˜ 5ë¡œ ë‚˜ëˆ„ì–´ì§„ë‹¤ë©´
    if n % 3 == 0 or n % 5 == 0: 
        
        # sì— nì˜ ê°’ì„ ë”í•œë‹¤.
        s+=n
        
# 3 + 5 + 6 + 9 = 23        
print(s)
```

**íŒŒì´ì¬ì˜ ê°ì²´ì§€í–¥ í”„ë¡œê·¸ë˜ë°(OOP)** ê¸°ëŠ¥ì„ ì´ìš©í•˜ì—¬ ì½”ë“œë¥¼ êµ¬ì„±í•˜ë©´ ì•„ë˜ì™€ ê°™ë‹¤.

```python
# ì¡°ê±´ì— ì¼ì¹˜í•˜ëŠ” ê°’ì´ ë“¤ì–´ê°ˆ ë¦¬ìŠ¤íŠ¸
m = list()

# nì— 1ì—ì„œ 9ê¹Œì§€ í• ë‹¹
for n in range(1, 10):
    
    # nì´ 3ìœ¼ë¡œ ë‚˜ëˆ„ì–´ì§€ê±°ë‚˜ 5ë¡œ ë‚˜ëˆ„ì–´ì§„ë‹¤ë©´
    if n % 3 == 0 or n % 5 == 0: 
        
        # ë¦¬ìŠ¤íŠ¸ mì— nì„ ì¶”ê°€í•œë‹¤.
        m.append(n)
        
print(sum(m))
```

ë‘ ë²ˆì§¸ ì˜ˆì œëŠ” ì²« ë²ˆì§¸ ì˜ˆì œì™€ ê²°ê³¼ëŠ” ê°™ì§€ë§Œ, ì§„í–‰í•˜ë©´ì„œ ì»¬ë ‰ì…˜ **ê°ì²´**ì¸ mì— ì¤‘ê°„ ê°’ì„ ëˆ„ì í•œë‹¤. ê³„ì‚°ì€ mê³¼ nì´ë¼ëŠ” ë³€ìˆ˜ì— ë“  ê°’ì— ì˜í•´ ê²°ì •ëœë‹¤.


í•˜ì§€ë§Œ í•¨ìˆ˜()ë‚˜ ê°ì²´, ë©”ì„œë“œ()ì™€ ê°™ì€ êµ¬ë¬¸(m.append(n), sum(m))ì„ ì‚¬ìš©í–ˆê¸° ë•Œë¬¸ì— ìˆœìˆ˜í•œ ê°ì²´ì§€í–¥ í”„ë¡œê·¸ë˜ë°ì´ ì•„ë‹ˆë¼ê³  (ì˜ëª») ì£¼ì¥í•  ìˆ˜ ìˆë‹¤. ê·¸ë ‡ë‹¤ë©´ ì¢€ ë” ê°ì²´ ëª¨ë¸ì„ ê·¹ë‹¨ì ìœ¼ë¡œ ì ìš©í•˜ì—¬ listì˜ í•˜ìœ„ í´ë˜ìŠ¤ë¥¼ ë§Œë“¤ì–´ sumì„ ì¶”ê°€í•  ìˆ˜ë„ ìˆë‹¤.

```python
class SummableList(list):
    
    # listì— í•˜ìœ„ í´ë˜ìŠ¤ë¥¼ ë§Œë“¤ì–´ sumì„ ì¶”ê°€í•œë‹¤.
    def sum(self):
        
        s = 0
        
        # ì´í„°ë ˆì´í„°(iterator): ë°˜ë³µì. ê°’ì„ ì°¨ë¡€ë¡œ êº¼ë‚¼ ìˆ˜ ìˆëŠ” ê°ì²´
        # ë°˜ë³µìëŠ” for i in range(100)ì´ë¼ëŠ” êµ¬ë¬¸ì´ ìˆìœ¼ë©´
        # 1~99ê¹Œì§€ ëª¨ë“  ê°’ì„ ë§Œë“¤ì–´ ë‘ì§€ ì•Šê³  í•„ìš”í•œ ê°’ë§Œ ì œë•Œ ë§Œë“œëŠ”ë°, ì´ê²ƒì„ í•˜ëŠ” ê²ƒì´ ì´í„°ë ˆì´í„°ë‹¤.
        # ë°ì´í„° ìƒì„±ì„ ë’¤ë¡œ ë¯¸ë£¨ë¯€ë¡œ ì´ëŸ° ë°©ì‹ì„ ì§€ì—° í‰ê°€(lazy evaluation)ë¼ê³  í•œë‹¤.
        
        # __iter__ ë©”ì„œë“œëŠ” ìš”ì†Œë¥¼ ì°¨ë¡€ë¡œ êº¼ë‚´ì¤€ë‹¤.
        for v in self.__iter__():
            s += v
            
        return s
```

ë³€ìˆ˜ mì„ list() ë©”ì„œë“œê°€ ì•„ë‹ˆë¼ SummableList í´ë˜ìŠ¤ë¡œ ì´ˆê¸°í™”í•œë‹¤ë©´, sum(m) ë©”ì„œë“œ ëŒ€ì‹  m.sum()ì„ ì‚¬ìš©í•  ìˆ˜ ìˆì„ ê²ƒì´ë‹¤. ì´ëŸ¬í•œ ì ì´ **íŒŒì´ì¬ì´ ê°ì²´ì§€í–¥ ì–¸ì–´ë¼ëŠ” ì‚¬ì‹¤**ì„ ëª…í™•íˆ ë³´ì—¬ì¤€ë‹¤. <U>ì „ìœ„ í‘œê¸°ë²•ì€ ë‹¨ì§€ êµ¬ë¬¸ìƒ í¸ì˜ë¥¼ ì œê³µí•  ë¿ì´ë‹¤.</U>

ìœ„ì—ì„œ ì´ì•¼ê¸°í•˜ê³ ì í•œ ë‚´ìš©ì´ ëª…ë ¹í˜• í”„ë¡œê·¸ë˜ë° ë°©ì‹ì´ ì˜ëª»ëë‹¤ëŠ” ë§ì´ ì•„ë‹ˆë‹¤. ì¤‘ìš”í•œ ê²ƒì€ ë‹¤ì–‘í•œ ê´€ì ì˜ ì „í™˜ì´ë‹¤. ì´ì œ í•¨ìˆ˜í˜• í”„ë¡œê·¸ë˜ë°ì„ ì‚´í´ë³¼ ê²ƒì´ë‚˜ ì´ ì˜ˆì œë¥¼ ê·¹ì ìœ¼ë¡œ ë” ì§§ê²Œ ë§Œë“¤ì–´ì£¼ê±°ë‚˜ ë” ë¹ ë¥´ê²Œ ë§Œë“¤ì–´ì£¼ì§€ëŠ” ëª»í•œë‹¤.

## í•¨ìˆ˜í˜• íŒ¨ëŸ¬ë‹¤ì„ ì‚¬ìš©í•˜ê¸°
ìœ„ ì˜ˆì œ(3ê³¼ 5ì˜ ë°°ìˆ˜ì˜ í•©)ë¥¼ í•¨ìˆ˜ ê´€ì ìœ¼ë¡œ ë³´ë©´, ë‘ ê°€ì§€ ë¶€ë¶„ìœ¼ë¡œ ë‚˜ëˆŒ ìˆ˜ ìˆë‹¤.

* ë¦¬ìŠ¤íŠ¸ **í•©ì„ ê³„ì‚°**
* ì˜ˆë¥¼ ë“¤ì–´ 3ì´ë‚˜ 5ì˜ ë°°ìˆ˜ì™€ ê°™ì€ ê°„ë‹¨í•œ ê²€ì‚¬ ì¡°ê±´ì„ í†µê³¼í•œ ê°’ì˜ ë¦¬ìŠ¤íŠ¸

1. ë¦¬ìŠ¤íŠ¸ í•©ì€ ê°„ë‹¨í•œ ì¬ê·€ì  ì •ì˜ë¡œ ë§Œë“¤ ìˆ˜ ìˆë‹¤.

```python
# ë¦¬ìŠ¤íŠ¸ í•©(ì¬ê·€)
def sum_seq(list):
    
    # ê¸°ì € ì¡°ê±´, listì˜ ê¸¸ì´ê°€ 0ì´ë©´ ì „ì²´ í•©ì€ 0ì„ ë°˜í™˜í•œë‹¤.
    if len(list) == 0: 
        return 0
    
    # ì‹œí€€ìŠ¤ì˜ í•©ì€ ì²« ë²ˆì§¸ ì›ì†Œì™€ ë‚˜ë¨¸ì§€ ì‹œí€€ìŠ¤ì˜ í•©ê³„ë¥¼ ë”í•œ ê°’ì´ë‹¤.
    return list[0] + sum(list[1:])
```

ì´ˆê¹ƒê°’ì„ ë¶„ë¦¬í•˜ê³  ë‚˜ë¨¸ì§€ ë¦¬ìŠ¤íŠ¸ í•©ê³„ë¥¼ ë”í•œ í˜•íƒœë¡œ ë§Œë“  ì´ìœ ëŠ” ì¶”í›„ ì„¤ëª…í•  ê²ƒì´ë‹¤.

2. ê²€ì‚¬ ì¡°ê±´ì„ í†µê³¼í•œ ê°’ì˜ ë¦¬ìŠ¤íŠ¸ë¥¼ ë„ì¶œí•˜ëŠ” í•¨ìˆ˜ë‹¤.

```python
# ìµœëŒ“ê°’ n, ì£¼ì–´ì§„ filter_func() í•¨ìˆ˜(Trueë‚˜ Falseë¥¼ ë°˜í™˜í•œë‹¤.), ì£¼ì–´ì§„ ê°’ v
def until(n, filter_func, v):
    
    # ê¸°ì € ì¡°ê±´
    # ì£¼ì–´ì§„ ê°’ vê°€ ìµœëŒ“ê°’ nê³¼ ê°™ì€ ê²½ìš° ë¹ˆ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°˜í™˜í•œë‹¤.(ë¦¬ìŠ¤íŠ¸ëŠ” ì•„ë˜ì„œ ê³„ì† ë¶™ì´ëŠ” ì¤‘)
    if v == n: 
        return []
    
    
    # filter_func() í•¨ìˆ˜ê°€ ì „ë‹¬ ë°›ì€ vë¥¼ í†µê³¼(True)ì‹œí‚¤ë©´
    if filter_func(v):
        
        # ì›ì†Œê°€ í•˜ë‚˜ë§Œ ë“¤ì–´ê°€ëŠ” ë§¤ìš° ì‘ì€ ë¦¬ìŠ¤íŠ¸ + ì¬ê·€ë¥¼ í†µí•´ v+1ì˜ ê²°ê³¼ë¥¼ ê³„ì†í•´ì„œ ë¶™ì—¬ì¤€ë‹¤.
        return [v] + until(n, filter_func, v+1)
    
    
    # ë§Œì•½ vê°’ì„ filter_funcê°€ ê±°ë¶€(False)í•œë‹¤ë©´, ë‹¤ìŒ v+1 ê°’ì„ ë„£ì€ filter_func()ë¥¼ í˜¸ì¶œí•œë‹¤.
    else:
        return until(n, filter_func, v+1)
```

ìœ„ì—ì„œ ë§Œë“  until()ë¥¼ ì´ìš©í•´ 3ì´ë‚˜ 5ì˜ ë°°ìˆ˜ë¥¼ ë§Œë“¤ ê²ƒì´ë‹¤. ë¨¼ì €, ê°’ì„ ê±¸ëŸ¬ë‚´ê¸° ìœ„í•œ filter_func() í•¨ìˆ˜ ì—­í• ì„ í•  ê°„ë‹¨í•œ lambda ê°ì²´ë¥¼ ë§Œë“¤ ê²ƒì´ë‹¤.

## í•¨ìˆ˜í˜• í”„ë¡œê·¸ë˜ë°ì˜ ê³ ì „ì ì¸ ì˜ˆì œ
$a_{i+1} = (a_{i}+n/a_{i})/2$ ìˆ˜ì—´ì„ ê³„ì‚°í•˜ëŠ” ì•„ë˜ ì˜ˆì œë¥¼ ì‚´í´ë³´ì

```python
# ë‚´ì¥ í•¨ìˆ˜ nextì™€ ê²¹ì¹˜ê¸° ë•Œë¬¸ì— next_ í•¨ìˆ˜ë¡œ í‘œê¸°
def next_(n, x):
    
    return (x+n/x) / 2
```

ë§¤ ë‹¨ê³„ ë°˜ë³µí• ìˆ˜ë¡ ê°’ ì‚¬ì´ì˜ ê±°ë¦¬ëŠ” ì¤„ì–´ë“¤ë©°, ì´ ìˆ˜ì—´ì€ ë¹ ë¥´ê²Œ a = n/aì¸ ê°’ìœ¼ë¡œ ìˆ˜ë ´í•œë‹¤. ì¦‰ $a = \sqrt{n}$ì´ë€ ì˜ë¯¸ë‹¤.

ë‹¤ìŒì€ ëª…ë ¹í–‰ í”„ë¡¬í”„íŠ¸ì—ì„œ ì´ í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•œ ê²°ê³¼ë‹¤.

```python
# n = 2ë¡œ ìˆ˜ì—´ì€ root 2ì— ìˆ˜ë ´í•  ê²ƒì´ë‹¤.
n = 2

f = lambda x: next_(n, x)

a0 = 1.0

[round(x, 4) for x in (a0, f(a0), f(f(a0)), f(f(f(a0))))] # ë’¤ëŠ” ì¬ê·€ì ì¸ ì‹œí€€ìŠ¤ë¡œ ê³„ì‚°í•œë‹¤.
```

n = 2 ì´ë¯€ë¡œ f() ë©”ì„œë“œëŠ” $\sqrt2$ë¡œ ìˆ˜ë ´í•˜ëŠ” lambdaë¡œ ì •ì˜ë˜ì—ˆë‹¤. a0 ì´ˆê¹ƒê°’ì€ 1.0ìœ¼ë¡œ ì„¤ì •í•´ ì‹œì‘í•˜ì˜€ìœ¼ë©° ì¬ê·€ì ì¸ í‰ê°€ë¥¼ í†µí•´ ê³„ì‚°í–ˆë‹¤.

ì•„ë˜ëŠ” ë¬´í•œ ìˆ˜ì—´ì„ ìƒì„±í•˜ëŠ” í•¨ìˆ˜ë‹¤.

```python
# next_() í•¨ìˆ˜(ìˆ˜ì—´ì˜ ê³„ì‚°), a0(ì´ˆê¹ƒê°’)ì„ ì¸ìë¡œ ë°›ì„ ê²ƒì´ë‹¤.
def repeat_(f, a):
    
    # ì´ˆê¹ƒê°’ì„ ë°˜í™˜
    yield a
    
    # ìˆ˜ì—´ì˜ ë‹¤ìŒ ê°’ì„ ê³„ì‚°í•´ì„œ ë°˜í™˜í•œë‹¤.(ì¬ê·€)
    for v in repeat_(f, f(a)):
        
        yield v 
```

```python
def within(ğœ–, iterable):
    
    # ë‚´ë¶€ í•¨ìˆ˜ ì •ì˜
    def head_tail(ğœ–, a, iterable):
        
        # ë‚´ì¥ í•¨ìˆ˜ nextëŠ” iteratorì—ì„œ ë‹¤ìŒ ê°’ì„ êº¼ë‚´ì¤€ë‹¤.
        # ìˆ˜ì—´ì—ì„œ aì˜ ë‹¤ìŒ ê°’ìœ¼ë¡œ bë¥¼ ìƒì„±í•œë‹¤.
        b = next(iterable)
        
        # ê¸°ì € ì¡°ê±´
        # a-bì˜ ì ˆëŒ“ê°’ì´ ğœ–(ê°ë‚´í•  ìˆ˜ ìˆëŠ” ì˜¤ì°¨ ë²”ìœ„)ë³´ë‹¤ ì‘ë‹¤ë©´ ìˆ˜ì—´ì„ ì¤‘ë‹¨í•˜ê³  bë¥¼ ë°˜í™˜í•œë‹¤.
        if abs(a-b) <= ğœ–:
            
            return b
        
        # ê¸°ì € ì¡°ê±´ì— í•´ë‹¹í•˜ì§€ ì•Šì„ ë•Œê¹Œì§€ ì¬ê·€
        # aì˜ ë‹¤ìŒ ê°’ì´ì—ˆë˜ bë¡œ í•¨ìˆ˜ë¥¼ ë‹¤ì‹œ ì§„í–‰í•œë‹¤.
        return head_tail(ğœ–, b, iterable)
    
    
    # ìœ„ì—ì„œ ì •ì˜í•œ head_tail í•¨ìˆ˜ë¥¼ ì‹œí–‰í•œë‹¤.
    return head_tail(ğœ–, next(iterable), iterable)
```
